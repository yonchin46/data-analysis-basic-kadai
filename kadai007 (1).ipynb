```python
#matplotlib,pandas,japanizeをインポート

import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib


#csvファイルをアップロードし変数に代入
category_df = pd.read_csv("sample_data/category.csv")
sample_df = pd.read_csv("sample_data/sample_pandas_6.csv")

#category_dfとsample_dfをマージ
df = pd.merge(sample_df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

# カテゴリー列のユニークな出現頻度をカウント
category_count = df['カテゴリー'].value_counts()

# 出現頻度を棒グラフで可視化
category_count.plot(kind='bar')
plt.xlabel('カテゴリー')
plt.ylabel('出現頻度')
plt.title('カテゴリー別の要素の出現頻度')

plt.show()
```

```python
# 商品番号列ごとにグループ化し、カテゴリー列を集計する。

quantity_count = df.groupby('商品番号')['カテゴリー'].count()

#商品番号列の商品番号ごとに、注文数の基本統計量を求める
quantity_desc = df.groupby('商品番号')['注文数'].describe()

print(quantity_count)
print(quantity_desc)
```
