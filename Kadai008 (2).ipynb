```python
#scikitlearnのdatasetsモジュールからload_wineをインポート
from sklearn.datasets import load_wine

#インポートしたload_wineをdataset変数に格納
dataset = load_wine()
#detasetを出力
dataset
```

```python
#データを見やすくするため、pandasのDataFreamを使用、表にまとまったかたちで出力される。
import pandas as pd
df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df.head()
```

```python
#ワインの種別を確認、ワインの種別は、datasetのtargetキーに含まれている。
dataset.target
```

```python
#サンプルデータの分割
X = dataset.data
y = dataset.target

#scikitlearnライブラリのmodel_serectionモジュールからtarain_test_split関数をインポート
from sklearn.model_selection import train_test_split
#xとyをトレーニングセットとテストセットに分割。test_size=0.3はデータの30％をテストセットに割り当てる。random_state=5は再現性を確保するためのシード＝データの分割が毎回同じになる。
train_test_split(X, y, test_size=0.3, random_state=5)


```

```python
#分割されたデータをそれぞれ、X_train,X_test,y_train,y_testに格納
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

#各データセットの形状（行数と列数）を出力（学習データ : テストデータ = 7 : 3に分割されている）
print(X.shape, X_train.shape, X_test.shape, y.shape, y_train.shape, y_test.shape)
```

```python
#Pythonのscikit-learnライブラリからDecisionTreeClassifierをインポート
from sklearn.ensemble import RandomForestClassifier

#RandomForestClassifierのインスタンスを作成し、model変数に格納
model = RandomForestClassifier(random_state=3)

```

```python
#機械学習モデルに学習させるコード、X_tarinはモデルが学習するデータ、y_trainはモデルが予測する対象
model.fit(X_train, y_train)
```

```python
#事前に学習されたmodelを使って、X_test（テストデータ）に対する予測を行う。その結果をy_predに格納し、出力。
y_pred = model.predict(X_test)
y_pred
```

```python
#目的変数のy-testのデータを確認
y_test
```

```python
#scikit-learnライブラリのmetricsモジュールからaccuracy_score関数をインポート
from sklearn.metrics import accuracy_score

#機械学習モデルの予測精度を評価、実際のラベル（y_test)と予測ラベル(y_pred)を比較して正しく予測されたサンプルの割合を計算する。
accuracy_score(y_test, y_pred)
```

```python
#scoreメソッドでの計算
model.score(X_test, y_test)
```
